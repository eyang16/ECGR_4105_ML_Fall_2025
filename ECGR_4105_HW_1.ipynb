```python
# Problem 1
# Normalize features for stability
df_norm = df.copy()
for col in ['X1', 'X2', 'X3']:
    df_norm[col] = (df_norm[col] - df_norm[col].mean()) / df_norm[col].std()

# Setup
Y = df_norm['Y'].values
variables = ['X1', 'X2', 'X3']
learning_rates = [0.1, 0.05, 0.01]
iterations = 1000
results = {}

# Cost function
def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    errors = predictions - y
    return (1 / (2 * m)) * np.sum(errors ** 2)

# Gradient descent
def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradient = (alpha / m) * X.T.dot(errors)
        theta -= gradient
        cost_history[i] = compute_cost(X, y, theta)
    return theta, cost_history

# Train models for each variable
for var in variables:
    X_raw = df_norm[var].values.reshape(-1, 1)
    X = np.hstack((np.ones((len(X_raw), 1)), X_raw))  # Add bias term
    best_loss = float('inf')
    best_theta = None
    best_lr = None
    best_history = None

    for lr in learning_rates:
        theta = np.zeros(2)
        theta, history = gradient_descent(X, Y, theta, lr, iterations)
        final_loss = history[-1]
        if final_loss < best_loss:
            best_loss = final_loss
            best_theta = theta
            best_lr = lr
            best_history = history

    results[var] = {
        'theta': best_theta,
        'loss': best_loss,
        'lr': best_lr,
        'history': best_history,
        'X': X_raw
    }

#  Plot
for var in variables:
    res = results[var]
    theta = res['theta']
    X_raw = res['X']
    Y_pred = theta[0] + theta[1] * X_raw

    plt.figure(figsize=(12,5))

    # Regression line
    plt.subplot(1,2,1)
    plt.scatter(X_raw, Y, color='red', label='Data')
    plt.plot(X_raw, Y_pred, color='green', label=f'Y = {theta[0]:.2f} + {theta[1]:.2f}*{var}')
    plt.xlabel(var)
    plt.ylabel('Y')
    plt.title(f'Regression Line ({var}, LR={res["lr"]})')
    plt.legend()

    # Loss curve
    plt.subplot(1,2,2)
    plt.plot(res['history'], color='blue')
    plt.xlabel('Iteration')
    plt.ylabel('Loss')
    plt.title(f'Loss Curve ({var})')

    plt.tight_layout()
    plt.show()

# Final report
print("\n Final Models:")
for var in variables:
    res = results[var]
    print(f"{var}: Y = {res['theta'][0]:.4f} + {res['theta'][1]:.4f} * {var} | "
          f"Loss = {res['loss']:.4f} | LR = {res['lr']}")

best_var = min(results, key=lambda k: results[k]['loss'])
print(f"\n Best explanatory variable: {best_var} with loss {results[best_var]['loss']:.4f}")
```

```python
# Problem 2
df_norm = df.copy()
for col in ['X1', 'X2', 'X3']:
    df_norm[col] = (df_norm[col] - df_norm[col].mean()) / df_norm[col].std()

X_raw = df_norm[['X1', 'X2', 'X3']].values
Y = df_norm['Y'].values
m = len(Y)

# Add bias term
X = np.hstack((np.ones((m, 1)), X_raw))

def compute_cost(X, y, theta):
    m = len(y)
    predictions = X.dot(theta)
    errors = predictions - y
    return (1 / (2 * m)) * np.sum(errors ** 2)

def gradient_descent(X, y, theta, alpha, iterations):
    m = len(y)
    cost_history = np.zeros(iterations)
    for i in range(iterations):
        predictions = X.dot(theta)
        errors = predictions - y
        gradient = (alpha / m) * X.T.dot(errors)
        theta -= gradient
        cost_history[i] = compute_cost(X, y, theta)
    return theta, cost_history

learning_rates = [0.1, 0.05, 0.01]
iterations = 1000
best_loss = float('inf')
best_theta = None
best_lr = None
best_history = None

for lr in learning_rates:
    theta = np.zeros(X.shape[1])
    theta, history = gradient_descent(X, Y, theta, lr, iterations)
    final_loss = history[-1]
    if final_loss < best_loss:
        best_loss = final_loss
        best_theta = theta
        best_lr = lr
        best_history = history

print(f"\nBest Learning Rate: {best_lr}")
print(f"Final Loss: {best_loss:.4f}")
print("Final Model:")
print(f"Y = {best_theta[0]:.4f} + {best_theta[1]:.4f}*X1 + {best_theta[2]:.4f}*X2 + {best_theta[3]:.4f}*X3")

plt.plot(best_history, color='blue')
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title(f'Loss Curve (LR={best_lr})')
plt.grid(True)
plt.show()

# Predicting new values
def predict(X_new, theta, df_stats):
    # Normalize input using training stats
    X_scaled = [(x - df_stats[col].mean()) / df_stats[col].std() for x, col in zip(X_new, ['X1', 'X2', 'X3'])]
    X_input = np.array([1] + X_scaled)  # add bias
    return X_input.dot(theta)

# Store stats for normalization
df_stats = df[['X1', 'X2', 'X3']]

# Predict
samples = [(1, 1, 1), (2, 0, 4), (3, 2, 1)]
for sample in samples:
    y_pred = predict(sample, best_theta, df_stats)
    print(f"Predicted Y for {sample}: {y_pred:.4f}")
```
